```python
from datetime import datetime, timedelta
from meteostat import Point, Daily, Hourly, Monthly
import pandas as pd
import matplotlib.pyplot as plt
plt.figSi
```

```python
raw_GL_df = pd.read_csv("df_us.csv")
raw_GL_df["pastYears"] = None
```

```python
for index, row in raw_GL_df.iterrows():
    date_start = datetime.strptime(row.event_date, '%Y-%m-%d %H:%M:%S')
    date_end = (date_start - timedelta(days=1825)).strftime('%Y-%m-%d %H:%M:%S')
    date_end = datetime.strptime(date_end, '%Y-%m-%d %H:%M:%S')
    lat = float(row.latitude)
    long = float(row.longitude)
    location = Point(lat, long, 0)

    data = Daily(location, date_end, date_start).fetch()
    if data.empty:
        continue
    else:
        weather_data = []
        for index2, day in data.iterrows():
            weather_data.append(day.values.flatten().tolist())
        raw_GL_df["pastYears"][index] = weather_data
```

```python
raw_GL_df.to_csv("us_landslide_past_5_years_days.csv")
```
